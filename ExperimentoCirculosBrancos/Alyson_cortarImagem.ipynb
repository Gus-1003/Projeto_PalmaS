## Bibliotecas e Ferramentas Utilizadas:

```python
import cv2
import numpy as np

from google.colab import drive
from google.colab.patches import cv2_imshow
```

## Importando imagem do Drive:

```python
drive.mount('/content/drive')
Img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/fig_control_1.jpg")
```

## Analise das dimensões da imagem:

*   800 pixels de largura x 800 pixels de comprimento
*   3 canais (Originalmente RGB)



```python
print(Img.shape) # Dimensões
```

```python
print(Img.item(0,0,2),Img.item(0,0,1),Img.item(0,0,0)) # Pegando a insentidade dos pixels em suas respectivas coordenadas
```

## Cortando Imagem:

```python
cv2_imshow(Img) 
```

```python
Img.itemset((0, 0, 2),255)
Img.itemset((0, 0, 1), 0)
Img.itemset((0, 0, 0), 0)
corte = Img[180:250, 250:315] # cortando uma parte da imagem
cv2_imshow(corte)
```

## Processo de sobreescrita da imagem:

```python
Imgcopy = Img.copy()
Imgcopy[30:100, 35:100] = corte # Colando a parte cortada sobre a original
cv2_imshow(Imgcopy) 
```
